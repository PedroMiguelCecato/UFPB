## Pandas

A partir do dataset `movies.csv` utilize o Pandas para responder as seguintes perguntas:

### Qual o gênero de filme mais lucrativo em todo o mundo?

```python
import pandas as pd

data = pd.read_csv('datasets/movies.csv')
display(data.head())
```

### Qual o diretor dirigiu mais filmes?

```python
diretores = data['Director'].groupby(data['Director']).count()
print(diretores.sort_values(ascending=False).head(10))
```

### Quais filmes tiverem os maiores arrecadações (Gross) em relação ao orçamento de produção (Budget)?

```python
lucro = data[['Title', 'US Gross', 'Production Budget']]
lucro['Profit'] = lucro['US Gross'] - lucro['Production Budget']
display(lucro.sort_values(by='Profit', ascending=False).head(10))
```
